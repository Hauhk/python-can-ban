```python
a = int(input("Nhập vào điểm 15p: "))
b = int(input("Nhập vào điểm kiểm tra giữa kỳ: "))
c = int(input("Nhập vào điểm kiểm tra giữa kỳ: "))

if a >=0 and b >=0 and c >=0 and a <=10 and b <= 10 and c <=10 :
    diemtb = (a + b*1.5 + c*2)/(1+1.5+2)
    if diemtb >= 8 : print("Đạt loại Giỏi môn Toán");
    elif 6.5 <= diemtb < 8 : print("Đạt loại Khá môn Toán")
    elif 5 <= diemtb < 6.5 : print("Đạt loại Trung Bình môn Toán")  
    elif 0 <= diemtb <5 : print("Đạt loại Yếu môn Toán")
else: print("Dữ liệu đầu vào sai")
```
