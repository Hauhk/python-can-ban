```python
import turtle

t = turtle.Turtle()
t.shape("turtle")
t.hideturtle()
t.speed(5)

t.penup()
t.goto(0,0)
t.pendown()
t.fillcolor('#f0ffff')
t.begin_fill()
t.circle(-120)
t.circle(80)
t.end_fill()

t.penup()
t.goto(30,100)
t.pendown()
t.fillcolor('black')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(-30,100)
t.pendown()
t.fillcolor('black')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(0,80)
t.pendown()
t.pencolor('orange')
t.fillcolor('orange')
t.begin_fill()
t.forward(15)
t.right(60)
t.backward(35)
t.end_fill()

t.penup()
t.goto(-30,60)
t.pendown()
t.pencolor('red')
t.pensize(4)
t.circle(40,120)

t.pensize(1)
t.pencolor('green')
t.penup()
t.goto(0,-50)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(0,-100)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(0,-150)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(0,-200)
t.pendown()
t.fillcolor('green')
t.begin_fill()
t.circle(5)
t.end_fill()

t.penup()
t.goto(-60,0)
t.pendown()
t.pencolor('red')
t.right(60)
t.pensize(20)
t.forward(120)

t.penup()
t.goto(30,0)
t.pendown()
t.right(90)
t.forward(30)

t.penup()
t.goto(0,150)
t.pendown()
t.pensize(1)
t.pencolor('green')
t.fillcolor('green')
t.begin_fill()
t.left(90)
t.forward(70)
t.left(135)
t.forward(60)
t.left(45)
t.forward(60)
t.left(45)
t.forward(60)
t.left(135)
t.forward(80)
t.end_fill()

t.penup()
t.goto(90,-50)
t.pendown()
t.pencolor('orange')
t.pensize(8)
t.left(45)
t.forward(55)
t.left(30)
t.forward(25)
t.penup()
t.backward(25)
t.pendown()
t.left(30)
t.forward(25)
t.penup()
t.backward(25)
t.pendown()
t.right(60)
t.forward(25)


t.penup()
t.goto(-90,-50)
t.pendown()
t.left(90)
t.forward(55)
t.right(30)
t.forward(25)
t.penup()
t.backward(25)
t.pendown()
t.right(45)
t.forward(25)
t.penup()
t.backward(25)
t.pendown()
t.left(90)
t.forward(25)

turtle.done()
```
